```
import engine
import model

from gensim.test.utils import datapath, get_tmpfile
from gensim.models import KeyedVectors
from gensim.scripts.glove2word2vec import glove2word2vec
from gensim.models.word2vec import Word2Vec
```

## Glove

```
glove_file = datapath('/home/jovyan/work/embeddings/glove/glove.6B.300d.txt')
word2vec_file = get_tmpfile('/home/jovyan/work/embeddings/word2vec/glove.6B.300d.txt')
_ = glove2word2vec(glove_file, word2vec_file)
word2vec = KeyedVectors.load_word2vec_format(word2vec_file)
```

```
e = engine.GameEngine(seed=123, expert=False, word2vec=word2vec)
e.play_game("computer", "human", "computer", "human", init=None)
```
